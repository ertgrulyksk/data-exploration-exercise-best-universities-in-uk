```python
print("\n'Top 5 Universities by UK Rank'")
print(bu.sort_values(by=['UK_rank'])[['University_name','Region']].head(5))
print("\n-------\n")
print("\n'Top 5 Universities by World Rank'")
print(bu.sort_values(by=['World_rank'])[['University_name','Region']].head(5))
print("\n-------\n")
```

```python
print("\n Private_University")
Private_University=bu.loc[bu['Control_type'] == 'Private'][['University_name','Region']]
print(Private_University)
print("\n-------\n")

print("\n\n Public_University")
Public_University=bu.loc[bu['Control_type'] == 'Public'][['University_name','Region']].head(5)
print(Public_University)

print("\n-------\n")
print("\n\n London_University")
London_University=bu.loc[bu['Region'] == 'London'][['University_name','World_rank']].head(5)
print(London_University)
```
